```python
#%%
from cProfile import label
from collections import Counter
import pandas as pd 
import matplotlib as mt 
import sklearn as sk
import numpy as np 
import sklearn.metrics as metrics
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
import seaborn as sb 
import matplotlib.pyplot as plt
import scipy.stats as stats 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn import preprocessing
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_curve, auc, precision_recall_curve
from statsmodels.stats.outliers_influence import variance_inflation_factor
from scipy.stats import norm
from sklearn.feature_selection import RFE
import statsmodels.api as sm

input_path = "/Users/ethan/Downloads/data 3.csv"
df = pd.read_csv(input_path)
print(df["Bankrupt?"])
bank_columns = list(df.columns)

def clean_col_names(col_name):
    col_name = (
        col_name.strip()
        .replace("?", "_")
        .replace("(", "_")
        .replace(")", "_")
        .replace(" ", "_")
        .replace("-", "_")
        .replace("__", "_")
        .replace("'", "")
        .lower()
    )
    return col_name
bank_columns = [clean_col_names(col_name) for col_name in bank_columns]
df.columns = bank_columns
print(df.columns)
print(df[df["bankrupt_"] == 1])
df.info()
print(df[df["bankrupt_"] == 1].value_counts)
scaler = preprocessing.MinMaxScaler()
df_scaled = scaler.fit_transform(df)
df = pd.DataFrame(df_scaled, columns= df.columns)
corr_mat = np.corrcoef(df,rowvar=False)
corr_mat = pd.DataFrame(corr_mat)
corr_mat.style.background_gradient(sb.light_palette('blue', as_cmap=True))
corr_heat = sb.heatmap(corr_mat, vmin=-1, vmax=1, center=True, annot=False)
plt.rcParams['figure.figsize'] = [40, 40]
plt.show()
df1 = df.loc[:,df.columns!='liability_assets_flag']
df2 = df1.drop(['bankrupt_'], axis=1)
df3 = pd.DataFrame(df2, columns= df2.columns)
df3['liability_assets_flag'] = df['liability_assets_flag']
df3['bankrupt_'] = df['bankrupt_']
df3['liability_assets_flag'] = df3['liability_assets_flag'].astype(object)
Y=df3["bankrupt_"]
X=df3.drop(['bankrupt_','net_income_flag', 'current_liability_to_equity','liability_assets_flag'], axis=1)
def calculate_vif_(X, thresh=10.0):
    variables = list(range(X.shape[1]))
    dropped = True
    while dropped:
        dropped = False
        vif = [variance_inflation_factor(X.iloc[:, variables].values, ix)
               for ix in range(X.iloc[:, variables].shape[1])]

        maxloc = vif.index(max(vif))
        if max(vif) > thresh:
            print('dropping \'' + X.iloc[:, variables].columns[maxloc] +
                  '\' at index: ' + str(maxloc))
            del variables[maxloc]
            dropped = True

    print('Remaining variables:')
    print(X.columns[variables])
    return X.iloc[:, variables]
X=X.pipe(calculate_vif_)
X.info()
corr_mat_1 = np.corrcoef(X,rowvar=False)
corr_mat_1 = pd.DataFrame(corr_mat_1)
corr_mat_1.style.background_gradient(sb.light_palette('blue', as_cmap=True))
corr_heat_1 = sb.heatmap(corr_mat_1, vmin=-1, vmax=1, center=True, annot=True)
plt.rcParams['figure.figsize'] = [20, 10]
plt.show()
print(corr_heat)
X_exog = sm.add_constant(X)
log_r_stats = sm.Logit(Y, X_exog).fit(method = 'lbfgs')
print(log_r_stats.summary())
resi_prob = [ ]
predictor_variable = [ ]
for i in range(len(X_exog.columns)):    
    log_residuals = sm.Logit(Y, X_exog.iloc[:,i]).fit(method = 'lbfgs')
    resi_prob.append(log_residuals.predict(X_exog.iloc[:,i]))
resi_prob = pd.DataFrame(resi_prob)
resi_prob = np.log(resi_prob / (1 - resi_prob))
resi_prob = pd.DataFrame(resi_prob)
resi_prob = resi_prob.transpose()
for i in range(len(X_exog)):
    predictor_variable.append(X_exog.loc[i])
fig, ax = plt.subplots(2,19, figsize=(250, 50))
for i in range(0,19):  
    ax[0,i].scatter(x = X_exog.iloc[:,i], y = resi_prob.iloc[:,i])
    ax[0,i].set_xlabel("Predictor")
    ax[0,i].set_ylabel("Log Odds")
for i in range(0,19):  
    ax[1,i].scatter(x = X_exog.iloc[:,(i+19)], y = resi_prob.iloc[:,(i+19)])
    ax[1,i].set_xlabel("Predictor")
    ax[1,i].set_ylabel("Log Odds")
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42, stratify=Y, shuffle=True)
log_r = LogisticRegression(class_weight='balanced')
log_r.fit(X_train,Y_train)
parameters = {"penalty": ['l1','l2'],
                  'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
                  'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}
randm_src = RandomizedSearchCV(estimator=log_r, param_distributions = parameters, n_iter = 10, n_jobs=-1,cv=5,scoring="f1")
randm_src=randm_src.fit(X_train,Y_train)
print(randm_src.best_params_)
log_r = randm_src.best_estimator_
y_train_pred = log_r.predict_proba(X_train)
y_train_pred_prob_lowerThres = y_train_pred[:,1] > 0.5
arg_train = {'y_true':Y_train, 'y_pred':y_train_pred_prob_lowerThres}
print(confusion_matrix(**arg_train))
print(classification_report(**arg_train))
y_test_pred = log_r.predict_proba(X_test)
y_test_pred_prob_lowerThres = y_test_pred[:,1] > 0.5
arg_train = {'y_true':Y_test, 'y_pred':y_test_pred_prob_lowerThres}
coefs = np.concatenate([log_r.coef_[0]])
Coef = pd.DataFrame(coefs, index=X.columns)
Intercept = pd.DataFrame(log_r.intercept_, index=['Intercept'])
intercept =pd.Series(['Intercept'])
Columns = pd.Series(X.columns)
Coef_Results = np.concatenate((Coef,Intercept), axis=0)
Columns_Names = Columns.append(intercept, ignore_index=True)
Column_Names = list(Columns_Names)
Coef_Results = pd.DataFrame(Coef_Results)
print(confusion_matrix(**arg_train))
print(classification_report(**arg_train))
log_rfe = RFE(log_r,n_features_to_select=15)
log_rfe=log_rfe.fit(X_train,Y_train)
mask = log_rfe.get_support()
X_train_rfe = log_rfe.transform(X_train)
X_test_rfe = log_rfe.transform(X_test)
features = pd.DataFrame({'features':list(X.iloc[:,0:].keys()), 'select':list(mask)})
features = list(features[features['select']==True]['features'])
print(log_rfe.estimator_.coef_)
log_r = log_r.fit(X_train_rfe,Y_train)
coefs = np.concatenate([log_r.coef_[0]])
Coef = pd.DataFrame(coefs, index=features)
Intercept = pd.DataFrame(log_r.intercept_, index=['Intercept'])
intercept =pd.Series(['Intercept'])
Columns = pd.Series(features)
Coef = np.concatenate((Coef,Intercept), axis=0)
Columns_Names = Columns.append(intercept, ignore_index=True)
Column_Names = pd.Index(list(Columns_Names))
Coef_Results = pd.DataFrame(Coef, index=Column_Names)
Coef_Results = pd.DataFrame(Coef_Results, index=Column_Names)
print(Column_Names)
y_train_pred = log_r.predict_proba(X_train_rfe)
y_train_pred_prob_lowerThres = y_train_pred[:,1] > 0.5
arg_train = {'y_true':Y_train, 'y_pred':y_train_pred_prob_lowerThres}
print(confusion_matrix(**arg_train))
print(classification_report(**arg_train))
y_test_pred = log_r.predict_proba(X_test_rfe)
y_test_pred_prob_lowerThres = y_test_pred[:,1] > 0.5
arg_train = {'y_true':Y_test, 'y_pred':y_test_pred_prob_lowerThres}
print(confusion_matrix(**arg_train))
print(classification_report(**arg_train))
```
